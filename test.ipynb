```python
from load_data import loadData
from logger import readYaml
from process_data import normData, countLabel, sampleMask
from process_data import slicsData, slicData, felzenszwalbData, lscData, seedsData
import numpy as np
import matplotlib.pyplot as plt

yaml_payh = 'dataset/data_info.yaml'

data_info_yaml = readYaml(yaml_payh)

data_name = 'Indian_pines'

data, label = loadData(data_info_yaml, data_name)
print('data shape:', data.shape)

ndata = normData(data)

counts, class_num = countLabel(label)
print('counts:', counts)
print('class_num:', class_num)

train_mask, test_mask = sampleMask(label, counts)
```

```python
def plots(seg_index, block_num):
    print('block num:',block_num)
    print('0?1:',np.min(seg_index))
    plt.figure()
    plt.imshow(seg_index)
    plt.show()
    plt.close()
```

```python
# slicData
seg_index1, block_num1 = slicData(ndata, 40, 0.2, 0.5)
plots(seg_index1, block_num1)
```
